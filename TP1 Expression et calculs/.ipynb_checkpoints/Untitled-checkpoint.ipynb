## A. Calculs dans l'interpréteur

### Expressions avec flottants et entiers
Toutes les données créées et modifiées par l'interpréteur sont typées. Nous nous interessons aux types principaux permettant de coder des nombres : `int`et `float`.

Les nombres **entiers** sont représentés par des objets de type entier : **`int`**. Exemple : `1`, `-1023`.

Les nombres **flottants** permettent de représenter une partie des nombres décimaux et rationnels. Le type associé est **`float`**. Ils comprennent systématiquement un point dans leur représentation textuelle pour les distinguer des entiers.

Les nombres flottants peuvent être définis avec une **notation scientifique** : $3,56 \times 10^{-3}$ s'écrit : `3.56e-3`


````python
print(1) # entier 1
print(-1023) # entier -1023
print(type(1)) # type de 1 : int
````

````python
print(1.) # flottant 1. Le point le distingue de l'entier de même valeur.
print(-1023.) # flottant -1023
print(3.56e-3) # définition avec notation scientifique
print(type(1.)) # type de 1. : float
````

Les **opérateurs** usuels **repectent les règles de priorité**. Ce sont : 
- `+` et `-` pour l'addition et la soustraction
- `*` et `/` pour la multiplication et la division
- `**` pour l'élévation à une puissance : `2**4` pour $2^4$

Pour la multiplication, l'addition et la soustraction, si l'un des 2 termes est un flottant, le résultat est un flottant.
La division donne toujours un flottant.

Les opérateurs spécifiques aux entiers sont :
- `//` pour la division entière : `7//3` donne l'entier 2
- `%` pour le modulo qui donne le reste de la division entière : `7%3` donne l'entier 1

### Affectation du résultat d'une expression
L'**affectation est une instruction** python qui consiste à donner un **nom** à un objet existant, obtenu par évalution d'une expression. Ce **nom** est une suite de caractères sans espace, sans accent (même si c'est possible), débutant par une lettre mais pouvant comprendre des chiffres ainsi que le caractère `_`. Les minimuscules et majuscules sont distinctes : `CPGE`, `Cpge`, `cpge` sont des noms différents.

Le résultat devient ainsi une variable que l'on peut afficher ou utiliser dans d'autres expressions.

L'**affectation** se fait avec le caractère `=` en **laissant un espace avant et après** par convention. L'instruction d'affectation ne doit pas être confondue avec l'égalité utilisée en mathématiques !
```python
leNom = expression # donne le nom leNom à l'objet obtenu par l'expression
```
Le nom de la variable doit être court mais aussi explicite. On utilisera `V` pour une vitesse par exemple.

Exemples : `x1`, `somme`, `mot_grand`, `grandMot`

````python
x = (3 + 1) / 45  # le nom x est donné au résultat de l'expression (3+1)/54, un flottant
print(x) # affichage de l'objet de nom x
print(x**2+1) # affichage du résultat d'une expression utilisant x
print(type(x)) # affichage du type de l'objet de nom x

y = x     # y et x sont deux noms d'un même objet
print(y)
````

### Exemple d'utilisation : dégradation d'une boisson dans l'organisme

La cinétique de dégradation d'une boisson alcoolisée dans l’organisme donne une évolution temporelle de sa concentration dans le sang régie par l'équation :

$C_2(t)=\frac{V_1}{V_2}C_{01} (1-e^{-k_1 t})-k_2 t$ en mol/L, $t$ en min,

avec :

- $C_{01}=0,90$ mol/L, concentration initiale dans l'estomac,
- $V_1=1,0$ L, le volume de boisson absorbé,
- $V_2=40$ L, le volume total de liquide et de sang dans le corps humain,
- $k_1=0,17$ /min, la constante de vitesse caractéristique de la 1ère phase,
- $k_2=71 \times 10^{-6}$ mol/L/min, la constante de vitesse caractéristique de la 2ème phase.

On rappelle qu'en France, une personne ne peut conduire que si le taux d’alcool dans son sang ne dépasse pas $C_{2max}=0,50$ g/L, soit $0,011$ mol/L.

### Exemple d'utilisation : circuit RLC
En physique, vous étudierez le comportement temporel d'un circuit électrique constitué d'un condensateur, d'une résistance et d'une bobine placés en série. Le tracé de la tension aux bornes du condensateur en fonction du temps est donné ci-dessous. Il s'agit d'un relevé expérimental.


Les caractéristiques du circuit (en USI) sont :
- résistance électrique de la résistance : $R=400 ~\Omega$
- capacité du condensateur : $C=105 \times 10^{-9}$ F
- inductance de la bobine : $L=0,9$ H.

La tension initiale est $U_0=9,8$ V.

L'évolution temporelle de ce circuit peut être modélisée par une équation différentielle de degré 2 dont la solution est donnée ci-dessous.

Avec : $z=\frac R 2 \sqrt{\frac C L} $ le facteur d'amortissement, $\omega_0=\frac 1 {\sqrt{LC}}$ la pulsation propre

et aussi $\omega_a = \omega_0 \sqrt {1-z^2}$ et $\phi = \arctan \frac{\sqrt{1-z^2}} z$ 

$$u(t) = U_0 \frac 1 {\sqrt {1-z^2}} e^{-\omega_0 z t} \sin \left(\omega_a t + \phi \right)$$

**A faire :**
- définir les constantes et commenter (caractère `#` en indiquant l'unité et/ou la grandeur physique)
- définir les variables intermédiaires associées à $z$, $\omega_0$, $\omega_a$ et $\phi$
- calculer la tension aux instants $t=0,005$ s et $t=0,0075$ s. Les résultats de ce modèle correspondent-ils aux résultats expérimentaux ?

*Remarque : dans un prochain TP, nous tracerons les courbes.*


````python
R = 400 # résistance
C = 105e-9 # capacité
L = 0.9 # inductance
U0 = 9.8 # tension initiale

w0 = 1 / sqrt(L*C) # pulsation propre
z = R / 2 * sqrt(C/L) # amortissement

wa = w0 * sqrt(1-z**2)
phi = atan(sqrt(1-z**2)/z)

t = 0.005
u = U0 * exp(-w0 * z * t) / sqrt(1-z**2) * sin(wa * t + phi)
print(t, u)

t = 0.0075
u = U0 * exp(-w0 * z * t) / sqrt(1-z**2) * sin(wa * t + phi)
print(t, u)
````

### Booléens, opérateurs et expression logique

Les booléens peuvent prendre uniquement deux valeurs : `True` ou `False`. Ils sont de type `bool`.

Les opérateurs de comparaison donnent comme résultats des booléns. Ces opérateurs sont :
- `==` et `!=` pour tester si deux objets sont égaux ou différents 
- `>`, `<`, `<=` et `>=` les opérateurs de comparaison.

````python
print(18>45)
print(32<40)
````

````python
a = 18 > 45
b = 32 < 40
print(a)
print(b)
print(type(a))
````

Les opérateurs s'appliquant sur les booléens sont `and`, `or` et `not`. Le résultat est aussi un booléen.

Les deux premiers opérateurs sont dits `paresseux` :
- `a and b` : si `a` est faux, quelque soit la valeur booléenne de `b` le résulat de l'expression est `False`. L'expression `b` n'a pas besoin d'être évaluée et n'est donc pas évaluée ;
- `a or b` : si `a` est vraie, quelque soit la valeur booléenne de `b` le résulat de l'expression est `True`. L'expression `b` n'est donc pas évaluée.
